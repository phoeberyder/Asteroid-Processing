```python
filename_lov = '/Users/user/Documents/VS Code/Radar/Asteroid-Processing/2025 FA22/2025FA22_lov_az_and_el_30_seconds.txt'
filename_deff = '/Users/user/Documents/VS Code/Radar/Asteroid-Processing/2025 FA22/2025FA22_deff_az_and_el_10_seconds.txt'
```

```python
data_lov = pd.read_csv(filename_lov, names=['Time', 'az', 'el'])
data_deff = pd.read_csv(filename_deff, names=['Time', 'az', 'el'])
```

```python
def biggest_difference(nums):
    differences = [abs(nums[i+1] - nums[i]) for i in range(len(nums)-1)]
    return max(differences)
```

```python
def offset(site1, site2):
    offsets = [abs(site1[i] - site2[i]) for i in range(len(site1))]
    return offsets
```

```python
diff = biggest_difference(data_lov['az'])
percent_across_beam = (diff/(0.03))*100

print("Percentage across beam travelled:",round(percent_across_beam, 1)) 
print(diff)
```
